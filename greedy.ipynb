```python
n = 1260
count = 0

coin_types =[500,100,50,10]
for coin in coin_types:
    count += n//coin
    n %=coin
print(count)
```

```python
# 큰수의 법칙
# 2022.06.02

n, m, k = map(int, input().split())
data = list(map(int, input().split()))

data.sort()
first = data[n-1]
second = data[n-2]

result = 0

while True:
    for i in range(k):
        if m == 0:
            break
        result += first
        m -=1
    if m == 0:
        break
    result += second
    m -=1
print(result)
```
