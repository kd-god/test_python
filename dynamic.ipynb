# 금광(2022.07.25)->내 풀이(위), 답(아래)
배운점:

dp테이블 사용 방법

어려웠던점:

단순히 3*3에 국한해서 생각하니 알고리즘을 다른 방향으로 짜버렸다. 오른쪽 아래, 오른쪽, 오른쪽 위로 움직일 수 있다는 생각을 어떤 방향으로든 가도 된다. 즉, 해당 열에 가장 큰 수를 고르면 된다란 단순한 생각을 했다. 그래서 틀렸다.

풀이법:

dp테이블을 이용한 다이나믹 프로그래밍을 활용하였다.

```python
for t in range(int(input())):
    n, m = map(int,input().split())
    graph = list(map(int,input().split()))
    index = 0
    dp = []
    for _ in range(n):
        dp.append(graph[index:index+m])
        index += m
    for j in range(1,m):
        for i in range(n):
            if i == 0:
                dp[i][j] = dp[i][j] + max(dp[i][j-1], dp[i+1][j-1], dp[i+2][j-1])
            elif i == 1:
                dp[i][j] = dp[i][j] + max(dp[i-1][j-1], dp[i][j-1], dp[i+1][j-1])
            else :
                dp[i][j] = dp[i][j] + max(dp[i-2][j-1], dp[i-1][j-1], dp[i][j-1])
    print(dp)
```

# 정수 삼각형(2022.07.25)

배운점:

금광과 같은 문제로 dp테이블을 효과적으로 사용하는 방법을 배움

어려웠던점:

점화식은 쉽게 세울수 있었으나, 이것을 실제로 표현하는게 살짝 까다로웠다.

풀이법:

dp테이블을 이용한 다이나믹 프로그래밍.

```python
n = int(input())
dp = []
for _ in range(n):
    dp.append(list(map(int,input().split())))
for i in range(1, n):
    for j in range(i+1):
        if j == 0:
            up_left = 0
        else:
            up_left = dp[i-1][j-1]
        if j == i:
            up = 0
        else:
            up = dp[i-1][j]
        dp[i][j] = dp[i][j] + max(up_left, up)
print(max(dp[n-1]))
```

# 퇴사(2022.07.25)

배운점:

점화식의 중요성에 대해 배웠다. 지금껏은 점화식을 따로 세우지 않고 무작정 문제를 풀었는데, 알고리즘은 설계의 과정이 중요하다는 것을 깨달았다.

어려웠던점:

처음엔 모든 경우의 수를 다음려고 했지만 dp테이블을 이용하면 훨씬 용이하게 문제를 풀수 있다는 것을 깨달았다. 그래서 뒤에서 계산하면서 dp테이블을 적용했다.

풀이법:

dp테이블에 뒤에서 부터 계산한 가치를 저장함으로써 풀어나갔다.

```python
n = int(input())
t= []
p =[]
dp = [0] * n
max_value = 0
for _ in range(n):
    x,y = map(int,input().split())
    t.append(x)
    p.append(y)
for i in range(n-1, -1, -1):
    time = t[i] + i
    if time <= n:
        dp[i] = max(p[i] + dp[time], max_value)
        max_value = dp[i]
    else:
        dp[i] = max_value
print(max_value)
```

# 병사 배치하기(2022.07.25)

배운점:

가장 긴 증가하는 부분 수열에 대해 배움.

어려웠던점:

단순 탐색으로 풀 수 있는 문제도 시간복잡도를 해결 하기 위해 점화식을 세운다는 점에서 헷갈렸다.

풀이법:

점화식을 사용하여 dp테이블을 생성했다.

```python
n = int(input())
array = list(map(int,input().split()))
array.reverse()
dp = [1] * 7
for i in range(1,n):
    for j in range(i):
        if array[j] < array[i]:
            dp[i] = max(dp[i], dp[j]+1)
print(n - max(dp))
```

# 못생긴수(2022.07.26)

배운점:

2,3,5만을 소인수로 갖는 수를 다루는 법에 대해 배웠다. 간단한 문제인듯 보였지만, dp테이블과 깔끔한 코드를 작성하기 까다로웠다.

어려웠던점:

다이나믹 프로그래밍 문제로서 dp테이블을 사용해야한다는 강박이 조금 어려웠다.

풀이법:

dp테이블 다이나믹 프로그래밍

```python
n = int(input())
dp = [0] * n
dp[0] = 1
i2 = i3 = i5 = 0
next2, next3, next5 = 2, 3, 5

for k in range(1,n):
    dp[k] = min(next2, next3, next5)
    if dp[k] == next2:
        i2+=1
        next2 = dp[i2] * 2
    if dp[k] == next3:
        i3 += 1
        next3 = dp[i3] * 3
    if dp[k] == next5:
        i5 += 1
        next5 = dp[i5] * 5
```

# 편집 거리(2022.07.26)

배운점:

리벤슈타인 알고리즘에 대해 배웠다.

어려웠던점:

기존 알고리즘을 이해하기가 조금 어려웠다. 몇번이고 왜 위쪽이 삭제의 의미고, 왼쪽 위가 교체의 의미고, 왼쪽이 삽입의 의미인지 고민했다. 결국엔 이해했다.

풀이법:

dp테이블을 이용한 리벤슈타인 알고리즘 정리

```python
def edit_dist(str1, str2):
    n = len(str1)
    m = len(str2)

    dp = [[0] *(m + 1) for _ in range(n+1)]

    for i in range(1, n+1):
        dp[i][0] = i
    for j in range(1, m+1):
        dp[0][j] = j

    for i in range(1, n+1):
        for j in range(1, m+1):
            if str1[i-1] == str2[j-1]:
                dp[i][j] = dp[i-1][j-1]
            else:
                dp[i][j] = 1 + min(dp[i][j-1], dp[i-1][j], dp[i-1][j-1])
    return dp[n][m]

str1 = input()
str2 = input()

print(edit_dist(str1,str2))
```
