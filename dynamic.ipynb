```python
# 다이나믹 프로그래밍
# 2022.07.07
# 한번 구현한 정보를 리스트에 저장해서 필요할 때 불러오는 방식 -> 메모제이션
```

```python
d = [0] * 100

def fibo(x):
    if x == 1 or x ==2:
        return 1
    if d[x] != 0:
        return d[x]
    d[x] = fibo(x-1) + fibo(x-2)
    return d[x]
fibo(99)
```

```python
x = int(input())

d = [0] * 3000

for i in range(2, x+1):
    d[i] = d[i-1] + 1
    if i % 2 == 0:
        d[i] = min(d[i],d[i//2] + 1)
    if i % 3 == 0:
        d[i] = min(d[i], d[i//3] + 1)
    if i % 5 ==0:
        d[i] = min(d[i], d[i//5] + 1)
print(d[x])
```

```python
# 개미전사
# 점화식 활용, 수학을 잘해야된다. 다이나믹 프로그래밍을 사용할 때는 항상 피보나치 수열을 떠올려야겠다.
# 2022.07.08
```

```python
n = int(input())
array  = list(map(int, input().split()))

d = [0] * 100
d[0] = array[0]
d[1] = max(array[0], array[1])

for i in range(2, n):
    d[i] = max(d[i-1], d[i-2]+ array[i])
print(d[n-1])
```

```python
n = int(input())
d = [0] * 100
d[0] = 1
d[1] = 3
for i in range(2, n):
    d[i] = (d[i-2] * 2 + d[i-1]) % 796796
print(d[n-1])
```

```python
# 효율적인 화폐구성
# 2022.07.08
```

```python
n, m = map(int, input().split())
array = []
for i in range(n):
    array.append(int(input()))

d = [10001] * (m+1)
d[0] = 0
for i in range(n):
    for j in range(array[i], m+1):
        d[j] = min(d[j], d[j - array[i]] + 1)

if d[m] == 10001:
    print(-1)
else :
    print(d[m])
```
