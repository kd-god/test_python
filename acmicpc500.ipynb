# 1번문제 - 브론즈 3


```python
a, b= map(int, input().split())
n = int(input())
min = b + n
if min >=60:
    new_hour = min // 60
    a += new_hour
    if a >=24:
        a -= 24
    min = min % 60

print(a, min)
```

# 3. 더하기 사이클 - 브론즈 1

```python
n = input()
if int(n) < 10:
        n = str(0)+str(n)
new_n = n
count = 0
while True:
    a = int(new_n[0])
    b = int(new_n[1])
    if a + b >=10:
        new_n = str(b) + (str(a+b))[1]
    else:
        new_n = str(b)+str(a+b)
    count += 1
    if new_n == n:
        break
print(count)
```

```python
graph = []
for _ in range(9):
    a = int(input())
    graph.append(a)
    
max_value_index = 0
max_value = 0

for i in range(1,len(graph)):
    if graph[i] > graph[max_value_index]:
        max_value_index = i
        max_value = graph[i]
        
print(max_value)
print(max_value_index+1)
```

```python
array = []
for _ in range(3):
    a = int(input())
    array.append(a)
    
result = array[0]
for i in range(1,len(array)):
    result *= array[i]
    
graph = [0] * 10
for i in range(len(graph)):
    graph[i] = list(str(result)).count(str(i))
    print(graph[i])
```

# 7. 셀프 넘버 - 실버5

리뷰:

list와 list는 -를 이용하여 차집합을 못구하지만 set와 set는 차집합을 -를 이용하여 차집합을 골라줄 수 있다는 점을 파악했고, sum(map(int,str(n)))을 이용하여 어떠한 숫자의 각자리의 합을 구할 수 있다는 사실을 파악했다.


---



***map(함수, list)가 주어지면 list 각각의 원소에 합수를 적용할 수 있다.***

```python
def d(n):
    n = n + sum(map(int,str(n)))
    return n

array = [False] * 10001

for i in range(1, 10001):
    if d(i) > 10000:
        continue
    array[d(i)] = True
    
for i in range(1,len(array)):
    if array[i] == False:
        print(i)
```

# 7. 셀프 넘버 - 실버5

리뷰:

문제는 맞았지만 쓸데없이 메모리 낭비를 했다. 왜냐하면 문제에서는 3자리수 만 고려한다했는데 난 모든 숫자를 고려하는 코드를 작성하다보니, 아직 시간 효율도 제대로 계산 못하는데 긴 코드를 작성했다. 다음부턴 이런 실수 안해야겠다. 


```python
n = int(input())
array = [False] * (n+1)
for i in range(1,n+1):
    if i <= 99:
        array[i] = True
    if i >= 100:
        array_new = list(map(int,str(i)))
        minus = array_new[1] - array_new[0]
        for j in range(1,len(array_new)-1):
            new_minus = array_new[j+1] - array_new[j]
            if minus == new_minus:
                array[i] = True
print(array.count(True))
```

# 8. 알파벳 찾기 - 브론즈5

리뷰:

풀이는 쉬웠으나, find 함수를 알게 되었다. 원하는 원소의 위치를 반환하고 존재하지 않을시 -1을 반환한다.

```python
n = input()
graph = []
for i in range(ord('a'),ord('z')+1):
    try:
        graph.append(n.index(chr(i)))
    except:
        graph.append(-1)
for i in graph:
    print(i, end=' ')
```

# 9. 단어 공부 - 브론즈1

리뷰:

문제역시 쉬웠으나, lower(), upper(), islower(), isupper()의 함수를 배웟다. lower()과 upper()는 원본에는 영향을 주지 않다는 다는 것을 알고 있어야하고, islower(), isupper()는 대문자 및 소문자 판단을 위해 true, false를 반환한다.

```python
n = input()
n = n.lower()
graph = [0] * (ord('z') - ord('a') + 1)
for i in n:
    graph[ord(i)-97] += 1
max_value = max(graph)
count = 0
for i in graph:
    if i == max_value:
        count += 1
if count >1:
    print('?')
else:
    a=chr(graph.index(max_value) + 97)
    print(a.upper())
```

# 크로아티아 변수 선언 - 실버5

리뷰:

replace함수를 쓰면되지만 나만의 알고리즘으로 풀고 싶었다. 하지만 테스트 케이스가 하나가 미스가 나면서 시간을 너무 낭비했다. 그래서 결국 replace로 해결하긴 했지만 dictionary의 정확한 사용법을 알게되었다.

```python
n = 'dz=ak'
dic = {'č':'c=', 'ć':'c-','dž':'dz=','đ':'d-','lj':'lj','nj':'nj','š':'s=','ž':'z='}
dic_keys = list(dic.keys())
dic_values = list(dic.values())
result = []
compare = ''
hi = ''
for i in n:
    compare += i
    if len(compare)>=3:
        hi += compare[1:len(compare)]
    for j in range(len(dic_values)):
        if hi == dic_values[j]:
            result.append(compare[0])
            result.append(hi)
            compare = ''
            hi = ''
            continue
        if compare == dic_values[j]:
            result.append(dic_keys[j])
            compare = ''
    

count = len(result) + len(compare)
print(count)
print(result)
print(compare)
```
