# 7. 셀프 넘버 - 실버5

리뷰:

list와 list는 -를 이용하여 차집합을 못구하지만 set와 set는 차집합을 -를 이용하여 차집합을 골라줄 수 있다는 점을 파악했고, sum(map(int,str(n)))을 이용하여 어떠한 숫자의 각자리의 합을 구할 수 있다는 사실을 파악했다.


---



***map(함수, list)가 주어지면 list 각각의 원소에 합수를 적용할 수 있다.***

```python
def d(n):
    n = n + sum(map(int,str(n)))
    return n

array = [False] * 10001

for i in range(1, 10001):
    if d(i) > 10000:
        continue
    array[d(i)] = True
    
for i in range(1,len(array)):
    if array[i] == False:
        print(i)
```

# 10. 크로아티아 변수 선언 - 실버5

리뷰:

replace함수를 쓰면되지만 나만의 알고리즘으로 풀고 싶었다. 하지만 테스트 케이스가 하나가 미스가 나면서 시간을 너무 낭비했다. 그래서 결국 replace로 해결하긴 했지만 dictionary의 정확한 사용법을 알게되었다.

```python
n = 'dz=ak'
dic = {'č':'c=', 'ć':'c-','dž':'dz=','đ':'d-','lj':'lj','nj':'nj','š':'s=','ž':'z='}
dic_keys = list(dic.keys())
dic_values = list(dic.values())
result = []
compare = ''
hi = ''
for i in n:
    compare += i
    if len(compare)>=3:
        hi += compare[1:len(compare)]
    for j in range(len(dic_values)):
        if hi == dic_values[j]:
            result.append(compare[0])
            result.append(hi)
            compare = ''
            hi = ''
            continue
        if compare == dic_values[j]:
            result.append(dic_keys[j])
            compare = ''
    

count = len(result) + len(compare)
print(count)
print(result)
print(compare)
```

```python
A,B,V = map(int, input().split())

if (V-B) % (A-B) == 0 :
    print((V-B) // (A-B))
else :
    print(((V-B) // (A-B)) +1)
```

```python
 n = int(input())
 group = 1
 parameter = 1
 while n > group:
     parameter += 1
     group += parameter
     
init = parameter

for i in range(1,n-(group-parameter)):
    init -= 1

if parameter % 2==0:
    print('{0}/{1}'.format((parameter+1)-init,init))
else:
    print('{0}/{1}'.format(init, (parameter+1)-init))
```

# 13. 설탕 배달 - 실버4

리뷰 :

단순 그리디 알고리즘 문제였는데, 장황하게 문제를 풀다가 결국엔 틀려 버렸다. 처음 밑의 알고리즘으로 풀다가 테스트 케이스는 맞추나 통과하지 못해서, 새로운 방법을 찾다가 그리디 알고리즘이란 것을 깨달았다. 문제를 풀기전 어떤 유형의 알고리즘을 적용할 수 있는지 파악하는 것도 중요하다는 생각을 했다.

** 문제점을 해결했다. 3이 0이되는 경우의 수를 배제 했기 때문에 오류가 떴었다.

```python
n = int(input())
parameter = n // 3
ex_count = int(1e9)
for i in range(parameter,-1, -1):
    count = 0
    three_result = n - (i * 3)
    five_process = three_result % 5
    if three_result == 0 or five_process== 0:
        five_result = three_result // 5
        count += (i + five_result)
        print(i, five_result)
        if count < ex_count:
            ex_count = count
if ex_count ==int(1e9):
    print(-1)
else:
    print(ex_count)
```
