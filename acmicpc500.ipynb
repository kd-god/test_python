# 1번문제 - 브론즈 3


```python
a, b= map(int, input().split())
n = int(input())
min = b + n
if min >=60:
    new_hour = min // 60
    a += new_hour
    if a >=24:
        a -= 24
    min = min % 60

print(a, min)
```

# 2번문제 - 브론즈 3

```python
array = list(map(int,input().split()))
array.sort()
confirm = 1
result = max(array) * 100
for i in range(len(array)-1):
    if array[i] == array[i+1]:
        confirm += 1
        if confirm == 3:
            result = 10000 + array[i] *1000
        elif confirm == 2:
            result = 1000 + array[i] * 100
print(result)
```

# 3. 더하기 사이클 - 브론즈 1

```python
n = input()
if int(n) < 10:
        n = str(0)+str(n)
new_n = n
count = 0
while True:
    a = int(new_n[0])
    b = int(new_n[1])
    if a + b >=10:
        new_n = str(b) + (str(a+b))[1]
    else:
        new_n = str(b)+str(a+b)
    count += 1
    if new_n == n:
        break
print(count)
```

# 4. 최댓값-브론즈3
(초기)

리뷰:

답은 맞았는데 계속 틀렸다고 하는걸 보니 속도에 문제가 있는 것 같아 위치 찾기 함수를 찾아보게 되었다.

```python
graph = []
for _ in range(9):
    a = int(input())
    graph.append(a)
    
max_value_index = 0
max_value = 0

for i in range(1,len(graph)):
    if graph[i] > graph[max_value_index]:
        max_value_index = i
        max_value = graph[i]
        
print(max_value)
print(max_value_index+1)
```

(개선)

리뷰:

index함수 잊지 않아야겠다. ***index함수는 원소의 위치를 반환 시켜준다.***

```python
graph = []
for _ in range(9):
    a = int(input())
    graph.append(a)
max_value = max(graph)
print(max_value)
print(graph.index(max_value)+1)
```

# 5. 숫자의 개수 - 브론즈2

리뷰:
리스트 내에서 사용할 수 있는 count 함수를 알게 되었다. **count 함수는 리스트 내에서 사용할 수 있다.**

```python
array = []
for _ in range(3):
    a = int(input())
    array.append(a)
    
result = array[0]
for i in range(1,len(array)):
    result *= array[i]
    
graph = [0] * 10
for i in range(len(graph)):
    graph[i] = list(str(result)).count(str(i))
    print(graph[i])
```

6. 나머지 - 브론즈2

리뷰 : set함수로 중복을 피할 수 있다.

```python
array = []
for _ in range(10):
    a = int(input())
    a %= 42
    array.append(a)
array_set = set(array)
count = 0
for _ in array_set:
    count +=1
```
