```python
n, m = map(int, input().split())
array2 = list(map(int, input().split()))

start = 0
end = max(array2)
result = 0

while (start <= end):
    mid = (start + end) // 2
    total = 0
    for i in array2:
        if i>mid:
            total += i - mid
    if total > m :
        start = mid + 1
    elif total == m :
        result = mid
        break
    else :
        end = mid - 1
print(result)
```

# 정렬된 배열에서 특정 수의 개수 구하기(2022.07.22)

배운점:

부등호 표시의 중요성을 배웠다. 구현하는 알고리즘을 정확히 인식하여 포함되는 원소인지 아닌지 정확히 알아야한다. 예를 들어 <와 <=의 차이를 인식함으로써 한층 더 성장할 수 있었다.

어려웠던잠:

<와 <= 이거때문에 한참의 시간을 날렸다.

풀이법:

이진탐색 알고리즘을 사용하였다. 최초의 타겟 원소의 위치, 마지막 타겟 원소의 위치를 구함으로써 알고리즘을 완성 시켰다.


```python
def first(array, start, end, target):
    if start > end:
        return None
    mid = (start + end) // 2
    if array[mid] == target and (target > array[mid-1] or mid == 0):
        return mid
    elif array[mid] < target:
        return first(array, mid + 1, end, target)
    else:
        return first(array, start, mid-1, target)

def last(array, start, end, target):
    if start > end:
        return None
    mid = (start + end) // 2
    if array[mid] == target and (target < array[mid+1] or mid == n-1):
        return mid
    elif array[mid] <= target:
        return last(array, mid + 1, end, target)
    else:
        return last(array, start, mid-1, target)

def solution(array, x):
    n = len(array)
    a = first(array, 0, n-1, x)
    if a == None:
        return 0
    b = last(array, 0, n-1, x)
    c = b-a+1
    return c

n,x = map(int,input().split())
array = list(map(int, input().split()))

count = solution(array, x)

if count == 0:
    print(-1)
else:
    print(count)
```

# 고정점 찾기(2022.07.24)

어려웠던점:

2진탐색 알고리즘을 숙지하고 있어서 어렵지 않게 해결 할 수 있었다.

풀이법:

2진탐색 알고리즘

```python
def binary_search(array, start, end):
    if start > end:
        return None
    mid = (start + end) // 2
    if array[mid] == mid:
        return mid
    elif array[mid] > mid:
        return binary_search(array, start, mid-1)
    else :
        return binary_search(array, mid + 1, end)

n = int(input())
array = list(map(int, input().split()))
x = binary_search(array, 0, n-1)
if x == None:
    print(-1)
else :
    print(x)
```

# 공유기 설치(2022.07.24)

배운점:

이진탐색을 응용하는 법을 배웠다. 

어려웠던점:

단순 이진탐색이 아니라, 응용의 버전이었기 때문에 어려웠다.

풀이법:

해결하지 못해서 답을 보고 해결 했다. 원소의 위치에 적용하는 것이 아닌, 원소간의 차이에 이진탐색을 적용하였다. 어려웠다.

```python
n, c = list(map(int,input().split()))

array = []
for _ in range(n):
    array.append(int(input()))
array.sort()

start = array[1] - array[0]
end = array[-1] - array[0]
result = 0

while(start <= end):
    mid = (start + end) // 2
    value = array[0]
    count = 1
    for i in range(1, n):
        if array[i] >= value + mid:
            value = array[i]
            count += 1
    if count >= c:
        start = mid + 1
        result = mid
    else:
        end = mid - 1
```

# 가사검색(2022.07.24)

배운점:

이진탐색법의 간략화 버전인 bisect에 대해 배울 수 있었다. 훨씬 간결하여 사용하기 용이하였다. 하지만 이진탐색의 알고리즘을 잊지 않고 언제 어디든 사용할 수 있도록 계속 외워둬야겠다.

어려웠던점:

문자열 문제에 대해 아직은 미숙하다보니, 문자열 관련 함수들을 자유롭게 사용하지 못했다. 실력적인 문제보다, 숙지의 문제이니 천천히 여러문제들을 해결해 나간다면 금방 숙지할 수 있을 것같다.

풀이법:

결국엔 bisect를 이용한 이진탐색이었다. 훨씬 간결한 이진탐색으로 문제를 해결 할 수 있었다. 그리고 queries 문자들을 길이별로 리스트를 따로 저장한 것은 좋은 접근법이었다 생각한다. 다음에는 bisect가 아닌 이진탐색 알고리즘을 실제 구현해 활용해 봐야겠다. 가장 중요한 점은 어떻게 문자열을 다루는지인 것 같다.

```python
from bisect import bisect_left, bisect_right

def count_by_range(a, left_value, right_value):
    right_index = bisect_right(a, right_value)
    left_index = bisect_left(a, left_value)
    return right_index - left_index

array = [[] for _ in range(10001)]
reversed_array = [[] for _ in range(10001)]

def solution(words, queries):
    answer = []
    for word in words:
        array[len(word)].append(word)
        reversed_array[len(word)].append(word[::-1])
    for i in range(10001):
        array[i].sort()
        reversed_array[i].sort()
    for q in queries:
        if q[0] != '?':
            res = count_by_range(array[len(q)], q.replace('?','a'), q.replace('?','z'))
        else:
            res = count_by_range(reversed_array[len(q)], q[::-1].replace('?','a'), q[::-1].replace('?','z'))
        answer.append(res)
    return answer
```
