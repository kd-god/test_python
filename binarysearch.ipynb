```python
# 이진 탐색
# 배열 내부의 데이터가 정렬되어 있어야만 사용할 수 있는 알고리즘임.
# 이 알고리즘은 시작점, 끝점, 중간점을 사용함.
# 2022.06.23
```

```python
def binary_search(array, target, start, end):
    if start > end:
        return None
    mid = (start+end)//2
    if array[mid] == target:
        return mid
    elif array[mid] > target:
        return binary_search(array, target, start, mid-1)
    else :
        return binary_search(array, target, mid + 1, end)
n, target = map(int, input().split())
array = list(map(int, input().split()))

result = binary_search(array, target, 0, n-1)
if result ==None:
    print("원소가 없슈")
else :
    print(result + 1)

    
```

```python
def binary_search(array, target, start, end):
    while start <= end:
        mid = (start + end) // 2
        if array[mid] == target:
            return mid
        elif array[mid] > target:
            end = mid - 1
        else :
            start = mid + 1
    return None
n = int(input())
array = list(map(int, input().split()))
array.sort()
m = int(input())
array2 = list(map(int, input().split()))

for i in array2:
    result = binary_search(array, i, 0, n - 1)
    if result == None:
        print('no', end = ' ')
    else :
        print('yes')
```

```python
# 떡볶이 최적 길이 찾기
# 데이터 입력 수가 크면 이진탐색부터 떠올린다
# 2022.06.27
```

```python
n, m = map(int, input().split())
array2 = list(map(int, input().split()))

start = 0
end = max(array2)
result = 0

while (start <= end):
    mid = (start + end) // 2
    total = 0
    for i in array2:
        if i>mid:
            total += i - mid
    if total > m :
        start = mid + 1
    elif total == m :
        result = mid
        break
    else :
        end = mid - 1
print(result)
```

# 정렬된 배열에서 특정 수의 개수 구하기(2022.07.22)

배운점:

부등호 표시의 중요성을 배웠다. 구현하는 알고리즘을 정확히 인식하여 포함되는 원소인지 아닌지 정확히 알아야한다. 예를 들어 <와 <=의 차이를 인식함으로써 한층 더 성장할 수 있었다.

어려웠던잠:

<와 <= 이거때문에 한참의 시간을 날렸다.

풀이법:

이진탐색 알고리즘을 사용하였다. 최초의 타겟 원소의 위치, 마지막 타겟 원소의 위치를 구함으로써 알고리즘을 완성 시켰다.


```python
def first(array, start, end, target):
    if start > end:
        return None
    mid = (start + end) // 2
    if array[mid] == target and (target > array[mid-1] or mid == 0):
        return mid
    elif array[mid] < target:
        return first(array, mid + 1, end, target)
    else:
        return first(array, start, mid-1, target)

def last(array, start, end, target):
    if start > end:
        return None
    mid = (start + end) // 2
    if array[mid] == target and (target < array[mid+1] or mid == n-1):
        return mid
    elif array[mid] <= target:
        return last(array, mid + 1, end, target)
    else:
        return last(array, start, mid-1, target)

def solution(array, x):
    n = len(array)
    a = first(array, 0, n-1, x)
    if a == None:
        return 0
    b = last(array, 0, n-1, x)
    c = b-a+1
    return c

n,x = map(int,input().split())
array = list(map(int, input().split()))

count = solution(array, x)

if count == 0:
    print(-1)
else:
    print(count)
```
