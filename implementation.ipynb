```python
n = int(input())
list_data = input().split()
x, y = 1, 1
dy = [-1, 1, 0, 0]
dx = [0, 0, -1, 1]
list_type = ['L', 'R', 'U', 'D']

for plan in list_data:
    for i in range(len(list_type)):
        if plan == list_type[i]:
            nx = x +dx[i]
            ny = y +dy[i]
    if nx < 1 or ny < 1 or nx > 5 or ny > 5:
        continue
    x, y = nx, ny
print(x, y)
```
