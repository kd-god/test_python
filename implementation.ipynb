```python
def solution(s):
    answer = len(s)
    for i in range(1,len(s)//2 + 1):
        now = s[0:i]
        new_list = ''
        count = 1
        for j in range(i, len(s), i):
            if now == s[j:j+i]:
                count += 1
            else :
                new_list += str(count) + now if count>= 2 else now
                now = s[j:j+i]
                count = 1
        new_list += str(count) + now if count>= 2 else now
        answer = min(answer, len(new_list))
    return answer
```

```python
def check(answer):
    for x, y, frame in answer:
        if frame == 0:
            if y == 0 or [x,y,1] in answer or [x,y-1,0] in answer or [x-1,y,1] in answer:
                continue
            return False
        elif frame == 1:
            if [x,y-1,0] in answer or [x+1, y-1, 0] in answer or ([x-1,y,1] in answer and [x+1, y, 1] in answer):
                continue
            return False
    return True
            

def solution(n, build_frame):
    answer = []
    for i in build_frame:
        x, y, frame, install = i
        if install == 1:
            answer.append([x,y,frame])
            if not check(answer):
                answer.remove([x,y,frame])
        if install == 0:
            answer.remove([x,y,frame])
            if not check(answer):
                answer.append([x,y,frame])
        
    return sorted(answer)
```
