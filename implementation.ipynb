```python
n = int(input())
list_data = input().split()
x, y = 1, 1
dy = [-1, 1, 0, 0]
dx = [0, 0, -1, 1]
list_type = ['L', 'R', 'U', 'D']

for plan in list_data:
    for i in range(len(list_type)):
        if plan == list_type[i]:
            nx = x +dx[i]
            ny = y +dy[i]
    if nx < 1 or ny < 1 or nx > 5 or ny > 5:
        continue
    x, y = nx, ny
print(x, y)
```

```python
# 나이트가 움직일 수 있는 경우의 수
# 2022.06.08
```

```python
input_data = input()
start_point_dy = int(input_data[1])
start_point_dx = int(ord(input_data[0])) - int(ord('a')) + 1
steps = [(-2,1), (-2,-1), (2,1), (2, -1), (1,2), (1,-2), (-1,2), (-1,-2)]
count = 0

for step in steps :
    dx = start_point_dx + step[0]
    dy = start_point_dy + step[1]
    if dx <= 0 or dx >8 or dy <= 0 or dy > 8:
        continue
    count +=1
print(count)
```

```python
n, m = map(int, input().split())
a, b ,direction = map(int, input().split())
array = [list(map(int, input().split())) for _ in range(n)]
d = [[0]*m for _ in range(n)]
d[a][b] = 1

d_row = [-1, 0, 1, 0]
d_column = [0, 1, 0, -1]


count = 1
turn_time = 0 

while True:
    n_column = b + d_column[direction]
    n_row = a + d_row[direction]
    if array[n_column][n_row] == 0 and d[n_column][n_row] == 0:
        d[n_column][n_row] = 1
        a = n_row
        b = n_column
        count += 1
        turn_time = 0
        
    else :
        direction += 1
        turn_time += 1
        if direction >=4 :
            direction = 0
        if turn_time ==3:
            n_column = b - d_column[direction]
            b_row = a-d_row[direction]
            if array[n_column][n_row] == 0:
                a = n_row
                b = n_column
            else:
                break
print(count)



    
```
