```python
input_data = input()
start_point_dy = int(input_data[1])
start_point_dx = int(ord(input_data[0])) - int(ord('a')) + 1
steps = [(-2,1), (-2,-1), (2,1), (2, -1), (1,2), (1,-2), (-1,2), (-1,-2)]
count = 0

for step in steps :
    dx = start_point_dx + step[0]
    dy = start_point_dy + step[1]
    if dx <= 0 or dx >8 or dy <= 0 or dy > 8:
        continue
    count +=1
print(count)
```

```python
# 게임 개발
# 2022.06.09
```

```python
n, m = map(int, input().split())
a, b ,direction = map(int, input().split())
array = [list(map(int, input().split())) for _ in range(n)]
d = [[0]*m for _ in range(n)]
d[a][b] = 1

d_row = [-1, 0, 1, 0]
d_column = [0, 1, 0, -1]


count = 1
turn_time = 0 

while True:
    n_column = b + d_column[direction]
    n_row = a + d_row[direction]
    if array[n_column][n_row] == 0 and d[n_column][n_row] == 0:
        d[n_column][n_row] = 1
        a = n_row
        b = n_column
        count += 1
        turn_time = 0
        
    else :
        direction += 1
        turn_time += 1
        if direction >=4 :
            direction = 0
        if turn_time ==3:
            n_column = b - d_column[direction]
            b_row = a-d_row[direction]
            if array[n_column][n_row] == 0:
                a = n_row
                b = n_column
            else:
                break
print(count)
```

```python
# 문자열 압추
# 2022.07.12
```

```python
def solution(s):
    answer = len(s)
    for i in range(1,len(s)//2 + 1):
        now = s[0:i]
        new_list = ''
        count = 1
        for j in range(i, len(s), i):
            if now == s[j:j+i]:
                count += 1
            else :
                new_list += str(count) + now if count>= 2 else now
                now = s[j:j+i]
                count = 1
        new_list += str(count) + now if count>= 2 else now
        answer = min(answer, len(new_list))
    return answer
```
