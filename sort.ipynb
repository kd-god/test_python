```python
# 선택정렬
# 가장 작은 데이터를 앞으로 보내는 과정을 N - 1번 수행함.
# 정렬 데이터 수가 10,000개 이상이면 정렬 속도가 급격히 느려짐.
# 2022.06.20
```

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]
for i in range(len(array)):
    min_indx = i
    for j in range(i+1, len(array)):
        if array[min_indx] > array[j]:
            min_indx = j
    array[i], array[min_indx] = array[min_indx], array[i]
print(array)
```

```python
# 삽입정렬
# 삽입정렬은 특정한 데이터가 적절한 위치에 들어가기 이전에, 그 앞까지의 데이터는 이미 정렬되어 있다고 가정함.
# 선택정렬과 달리 모든 데이터를 비교하는 것이 아니라 효율적임.
# 즉, 삽입 정렬은 필요할 때만 위치를 바꾸므로 데이터가 거의 정렬 되어 있을 때 훨씬 효율적임.
# 삽입 정렬은 두번째 데이터부터 시작함.
# 정렬 되어 있다면 퀵 알고리즘보다 훨씬 빠름.
# 정렬이 되어 있지 않다면 O(N^2)이지만 정렬되어있을 때는 1등임.
# 2022.06.20
```

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]
# break, continue를 쓰면 완전히 나가는 것이 아닌 그 loop만 나감.
for i in range(1,len(array)):
    for j in range(i, 0, -1):
        if array[j] < array[j-1]:
            array[j], array[j-1] = array[j-1], array[j]
        else :
            continue
print(array)
```

```python
array = [5,7,9,0,3,1,6,2,4,8]

def quick_sort(array, start, end):
    if start >= end:
        return array
    pivot = start
    left = start + 1
    right = end
    while left <= right:
        while left <= end and array[left] <=array[pivot]:
            left += 1
        while right > start and array[right] >= array[pivot]:
            right -= 1
        if left > right:
            array[right], array[pivot] = array[pivot], array[right]
        else:
            array[left], array[right] = array[right], array[left]
    quick_sort(array, start, right - 1)
    quick_sort(array, right + 1, end)

quick_sort(array,0,len(array)-1)
print(array)
```

```python
array3 = [5,7,9,0,3,1,6,2,4,8]

def quick_sort(array):
    if len(array) <= 1:
        return array

    pivot2 = array[0]
    tail = array[1:]

    left_side = [x for x in tail if x <= pivot2]
    right_side = [x for x in tail if x > pivot2]

    return quick_sort(left_side) + [pivot2] + quick_sort(right_side)

print(quick_sort(array3))
```

# 실패율(2022.07.21) - 초기코드

간단한 리뷰:

간단한 문제였지만 다수의 예외때문에 애를 먹었다. 결국 풀어내긴 했지만 좀 더 깔끔한 코드를 필요로 할 것 같다.

```python
def solution(N, stages):
    answer = []
    data = [0] * (N+1)
    for i in stages:
        if i > N:
            continue
        data[i] += 1
        
    whole_person = len(stages)
    
    for i in range(1, N+1):
        answer.append((i, data[i]/whole_person))
        whole_person -= data[i]
        if whole_person ==0:
            whole_person = 1
            
    answer.sort(key = lambda x : (-x[1],x[0]))
    answer=[i[0] for i in answer]
    return answer
```

# 실패율(2022.07.21) - 개선코드

배운점:

코드를 정갈하게 짜야 겠다는 생각이 들었다. 기존에 내가 짠 코드는 풀이를 위해 만들어진 알고리즘이라 하면, 답지의 코드는 내 코드를 좀더 일반화 했다.

어려웠던점;

.count와 같은 파이썬 모듈을 좀 더 공부 해야겠다.


```python
def solution(N, stages):
    answer = []
    length = len(stages)

    for i in range(1, N + 1):
        count = stages.count(i)

        if length == 0:
            fail = 0
        else :
            fail = count / length

        answer.append((i, fail))
        length -= count
    answer = sorted(answer, key = lambda t: t[1], reverse = True)

    answer = [i[0] for i in answer]
    return answer
```

# 카드 정렬하기(2022.07.22)

배운점 :

우선순위 큐에 대해 배웠다.
원소가 작은 순서부터 작업을 한다.

어려웠던점:

어려운점은 없었다.

풀이법:

사실 우선순위 큐를 사용하지 않아도 되지만, heapq가 속도가 굉장히 빠르다는 점에서 자주 사용해야겠다는 생각을 함.



```python
import heapq
n = int(input())
heap =[]
for _ in range(n):
    data = int(input())
    heapq.heappush(heap,data)

answer = 0

while len(heap) != 1:
    one = heapq.heappop(heap)
    two = heapq.heappop(heap)
    result = one + two
    answer += result
    heapq.heappush(heap,result)
```
