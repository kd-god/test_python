```python
# 선택정렬
# 가장 작은 데이터를 앞으로 보내는 과정을 N - 1번 수행함.
# 정렬 데이터 수가 10,000개 이상이면 정렬 속도가 급격히 느려짐.
# 2022.06.20
```

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]
for i in range(len(array)):
    min_indx = i
    for j in range(i+1, len(array)):
        if array[min_indx] > array[j]:
            min_indx = j
    array[i], array[min_indx] = array[min_indx], array[i]
print(array)
```

```python
# 삽입정렬
# 삽입정렬은 특정한 데이터가 적절한 위치에 들어가기 이전에, 그 앞까지의 데이터는 이미 정렬되어 있다고 가정함.
# 선택정렬과 달리 모든 데이터를 비교하는 것이 아니라 효율적임.
# 즉, 삽입 정렬은 필요할 때만 위치를 바꾸므로 데이터가 거의 정렬 되어 있을 때 훨씬 효율적임.
# 삽입 정렬은 두번째 데이터부터 시작함.
# 정렬 되어 있다면 퀵 알고리즘보다 훨씬 빠름.
# 정렬이 되어 있지 않다면 O(N^2)이지만 정렬되어있을 때는 1등임.
# 2022.06.20
```

```python
array = [7, 5, 9, 0, 3, 1, 6, 2, 4, 8]
# break, continue를 쓰면 완전히 나가는 것이 아닌 그 loop만 나감.
for i in range(1,len(array)):
    for j in range(i, 0, -1):
        if array[j] < array[j-1]:
            array[j], array[j-1] = array[j-1], array[j]
        else :
            continue
print(array)
```

```python
# 퀵 정렬
# 2022.06.21
```

```python
array = [5,7,9,0,3,1,6,2,4,8]

def quick_sort(array, start, end):
    if start >= end:
        return array
    pivot = start
    left = start + 1
    right = end
    while left <= right:
        while left <= end and array[left] <=array[pivot]:
            left += 1
        while right > start and array[right] >= array[pivot]:
            right -= 1
        if left > right:
            array[right], array[pivot] = array[pivot], array[right]
        else:
            array[left], array[right] = array[right], array[left]
    quick_sort(array, start, right - 1)
    quick_sort(array, right + 1, end)

quick_sort(array,0,len(array)-1)
print(array)
```

```python
# 파이썬 장점을 살린 퀵정렬
# 2022.06.21
```

```python
array3 = [5,7,9,0,3,1,6,2,4,8]

def quick_sort(array):
    if len(array) <= 1:
        return array

    pivot2 = array[0]
    tail = array[1:]

    left_side = [x for x in tail if x <= pivot2]
    right_side = [x for x in tail if x > pivot2]

    return quick_sort(left_side) + [pivot2] + quick_sort(right_side)

print(quick_sort(array3))
```

```python
# 계수정렬
# 대부분의 정렬문제는 기본정렬 라이브러리를 사용하고, 아니면 계수정렬을 사용해라.
# 2022.06.22
```

```python
array = [7,5,9,0,3,1,6,2,9,1,4,8,0,5,2]
count = [0] * (max(array)+1)

for i in range(len(array)):
    count[array[i]] += 1

for i in range(len(count)):
    for j in range(count[i]):
        print(i, end=' ')
```
