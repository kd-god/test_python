```python
import heapq
import sys
input = sys.stdin.readline
INF = int(1e9)

n, m = map(int, input().split())
start = int(input())
graph = [[] for i in range(n+1)]
distance = [INF] * (n+1)

for _ in range(m):
    a, b, c = map(int, input().split())
    graph[a].append((b,c))
def dijkstra(start):
    q = []
    heapq.heappush(q, (0, start))
    distance[start] = 0
    while q:
        dist, now = heapq.heappop(q)
        if distance[now] < dist:
            continue
        for i in graph[now]:
            cost = dist + i[1]
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))
dijkstra(start)

for i in range(1, n+1):
    if distance[i] == INF:
        print('INFINITY')
    else:
        print(distance[i])
```

어려웠던점:

알고리즘 이해는 쉬우나 코드로 짜는 것이 살짝 난해했다.

풀이법:

서로소 집합 알고리즘을 통해서 여행지를 비교할때, 노드번호가 낮은 여행지에 find_parent함수와  union_parent함수를 이용하여 노드번호가 낮은 여행지를 부모노드를 선정함으로써 문제를 해결했다.

```python
n, m = map(int, input().split())
graph = [list(map(int,input().split())) for _ in range(n)]
plan = list(map(int, input().split()))
parent = []
for i in range(n+1):
    parent.append(i)

def find_parent(parent, x):
    if parent[x] != x:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union_parent(parent, a, b):
    a = find_parent(parent, a)
    b = find_parent(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

for i in range(n):
    for j in range(n):
        if graph[i][j] == 1:
            union_parent(parent,i+1,j+1)

result = False

for i in range(m-1):
    if find_parent(parent,plan[i]) == find_parent(parent, plan[i+1]):
        result = True
    else :
        result = False
print(result)
```

# 행성터널(2022.07.31)

배운점:

서로소 알고리즘을 응용했다. 3차원 데이터 이기에 3차원 데이터를 초기화 하는 시도들을 하려 했는데, 추후 데이터가 더 커질때면 문제가 생길 수 있기에 새로운 해결 방법을 찾게되었다.

어려웠던점:

시간 문제 해결법, 즉 기존 서로소 알고리즘을 응용하는 점에서 굉장히 헷갈렸다.

풀이법:

3차원 데이터를 구축하지 않고, x 데이터는 x 데이터, y 데이터는 y 데이터, z 데이터는 z 데이터끼리 계산함으로써 어차피 작은 데이터부터 확인하면 되니깐 문제를 해결해나갔다. 서로소 집합 알고리즘 + 크루스칼 알고리즘을 사용했다.

```python
n = int(input())
x = []
graph = []
y = []
z = []
parent = []
for i in range(n+1):
    parent.append(i)

def find_parent(parent,x):
    if parent[x] != x:
        parent[x] = find_parent(parent, parent[x])
    return parent[x]

def union_parent(parent, a, b):
    a = find_parent(parent, a)
    b = find_parent(parent, b)
    if a < b:
        parent[b] = a
    else:
        parent[a] = b

for i in range(1, n+1):
    a, b, c = map(int, input().split())
    x.append((a, i))
    y.append((b, i))
    z.append((c, i))

x.sort()
y.sort()
z.sort()

for i in range(n-1):
    graph.append((abs(x[i+1][0] - x[i][0]), x[i][1], x[i+1][1]))
    graph.append((abs(y[i+1][0] - y[i][0]), y[i][1], y[i+1][1]))
    graph.append((abs(z[i+1][0] - z[i][0]), z[i][1], z[i+1][1]))

graph.sort()
result = 0

for i in graph:
    cost, a, b = i
    if find_parent(parent, a) != find_parent(parent, b):
        union_parent(parent, a, b)
        result += cost
print(result)
```
