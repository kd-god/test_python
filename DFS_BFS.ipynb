```python
from collections import deque
queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue)
queue.reverse()
print(queue)

# list자료형으로 바꾸기
print(list(queue))
```

```python
# 재귀함수
# 2022.06.14
```

```python
def factorial_recursive(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    print(result)

def factorial_recursive2(n):
    if n == 1:
        return 1
    return n *factorial_recursive2(n-1)

print(factorial_recursive2(5))
factorial_recursive(5)        
```

```python
# DFS/BFS
# 인접행렬 표현 방식
# 2022.06.14
```

```python
INF = 9999999
graph =[
        [0,7,5],
        [7,0,INF],
        [5, INF, 0]
]
print(graph)
```

```python
# DFS/BFS
# 인접 리스트 표현 방식
# 2022.06.14
```

```python
graph = [[] for _ in range(3)]
graph[0].append((1,7))
graph[0].append((1,7))
graph[1].append((2,3))
graph[2].append((4,5))

print(graph)
print(graph[0][1])
```
