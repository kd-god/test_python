```python
def factorial_recursive(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    print(result)

def factorial_recursive2(n):
    if n == 1:
        return 1
    return n *factorial_recursive2(n-1)

print(factorial_recursive2(5))
factorial_recursive(5)        
```

```python
visited = [0] * 9

def dfs(graph, v, visited):
    visited[v] = True
    print(v, end =' ')
    for i in graph[v]:
        if visited[i] == False:
            dfs(graph, i, visited)
graph=[[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7]]
dfs(graph, 1, visited)
```

```python
# BFS는 너비우선 탐색. 큐 사용. 큐는 From collections import deque를 사용
# 큐는 선입선출 방식임
# 2022.06.15
```

```python
from collections import deque
visited = [0] *9
def bfs(graph, v, visited):
    visited[v] = True
    queue = deque([v])
    while queue:
        out_data = queue.popleft()
        print(out_data, end = ' ')
        for i in graph[out_data]:
            if not visited[i]:
                visited[i] = True
                queue.append(i)
bfs(graph, 1, visited)

```

```python
n, m = map(int, input().split(' '))
graph = [list(map(int, input())) for _ in range(n)]
count = 0
def dfs(x,y):
    if x <= -1 or x >= n or y <= -1 or y >=n:
        return False
    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs(x-1, y)
        dfs(x+1, y)
        dfs(x, y - 1)
        dfs(x, y + 1)
        return True
    return False

for i in range(n):
    for j in range(m):
        if dfs(i,j) == True:
            count +=1
print(count)
```

```python
from collections import deque

n, m = map(int, input().split())
graph = [list(map(int, input())) for _ in range(n)]
dx = [0,0,-1,1]
dy = [1,-1,0,0]
 
def bfs(graph, x, y):
    queue = deque()
    queue.append((x,y))
    while queue:
        x, y = queue.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if nx < 0 or ny < 0 or nx >= n or ny >= m:
                continue
            if graph[nx][ny] == 0:
                continue
            if graph[nx][ny] == 1:
                queue.append((nx,ny))
                graph[nx][ny] = graph[x][y] + 1

    return graph[n-1][m-1]
print(bfs(graph,0,0))

```

```python
n, m = map(int, input().split())
data = []
temp = [[0] * m for _ in range(n)]

for _ in range(n):
    data.append(list(map(int,input().split())))

dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

result = 0

def virus(x,y):
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if nx >= 0 and ny >=0 and nx < n and ny < m:
            if temp[nx][ny] ==0:
                temp[nx][ny] = 2
                virus(nx,ny)
def get_score():
    score = 0
    for i in range(n):
        for j in range(m):
            if temp[i][j] == 0:
                score += 1
    return score

def dfs(count):
    global result
    if count == 3:
        for i in range(n):
            for j in range(m):
                temp[i][j] = data[i][j]
        for i in range(n):
            for j in range(m):
                if temp[i][j] == 2:
                    virus(i,j)
        result = max(result, get_score())
        return

    for i in range(n):
        for j in range(m):
            if data[i][j] == 0:
                data[i][j] = 1
                count += 1
                dfs(count)
                data[i][j] = 0
                count -= 1
dfs(0)
print(result)
```

```python
from collections import deque
n, k = map(int,input().split())

graph = []
data = []

for i in range(n):
    graph.append(list(map(int,input().split())))
    for j in range(len(graph[i])):
        if graph[i][j] != 0:
            data.append((graph[i][j],0, i,j))

target_s,target_x,target_y = map(int,input().split())

dx = [1,0,-1,0]
dy = [0,1,0,-1]

data.sort()
queue = deque(data)

while queue:
    virus, s, x, y = queue.popleft()
    if s == target_s:
        break
    for i in range(4):
        nx = x + dx[i]
        ny = y + dy[i]
        if ny < n and nx < n and ny>=0 and nx >=0:
            if graph[nx][ny] ==0:
                graph[nx][ny]= virus
                queue.append((virus, s+1, nx, ny))
print(graph[target_x-1][target_y-1])
```
