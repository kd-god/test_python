```python
def factorial_recursive(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    print(result)

def factorial_recursive2(n):
    if n == 1:
        return 1
    return n *factorial_recursive2(n-1)

print(factorial_recursive2(5))
factorial_recursive(5)        
```

```python
# DFS는 스택자료형 사용
# 스택은 선입후출 방식임
# 2022.06.15
```

```python
visited = [0] * 9

def dfs(graph, v, visited):
    visited[v] = True
    print(v, end =' ')
    for i in graph[v]:
        if visited[i] == False:
            dfs(graph, i, visited)
graph=[[],[2,3,8],[1,7],[1,4,5],[3,5],[3,4],[7],[2,6,8],[1,7]]
dfs(graph, 1, visited)
```

```python
# BFS는 너비우선 탐색. 큐 사용. 큐는 From collections import deque를 사용
# 큐는 선입선출 방식임
# 2022.06.15
```

```python
from collections import deque
visited = [0] *9
def bfs(graph, v, visited):
    visited[v] = True
    queue = deque([v])
    while queue:
        out_data = queue.popleft()
        print(out_data, end = ' ')
        for i in graph[out_data]:
            if not visited[i]:
                visited[i] = True
                queue.append(i)
bfs(graph, 1, visited)

```

```python
# 음료수 얼려먹기
# 2022.06.15
```

```python
n, m = map(int, input().split(' '))
graph = [list(map(int, input())) for _ in range(n)]
count = 0
def dfs(x,y):
    if x <= -1 or x >= n or y <= -1 or y >=n:
        return False
    if graph[x][y] == 0:
        graph[x][y] = 1
        dfs(x-1, y)
        dfs(x+1, y)
        dfs(x, y - 1)
        dfs(x, y + 1)
        return True
    return False

for i in range(n):
    for j in range(m):
        if dfs(i,j) == True:
            count +=1
print(count)
```
