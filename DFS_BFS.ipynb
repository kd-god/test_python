```python
p = '()))((()'

def balanced_word(p):
    count = 0
    for i in range(len(p)):
        if p[i] =='(':
            count += 1
        else:
            count -= 1
        if count == 0:
            return i

def right_word(p):
    count = 0
    for i in range(len(p)):
        if p[i] == '(':
            count += 1
        else:
            if count == 0:
                return False
            else:
                count -=1
    return True


def solution(p):
    answer = ''
    if p == '':
        return answer
    index = balanced_word(p)
    u = p[:index+1]
    v = p[index+1:]
    if right_word(u):
        answer += u
        answer += solution(v)
        return answer
    else:
        answer += '('
        answer += solution(v)
        answer += ')'
        u = answer[1:-1]
        for i in u:
            if i =='(':
                i = ')'
            else:
                i = '('
    return answer + u
        
print(solution(p))
```

```python
from itertools import permutations
from collections import deque

n = int(input())
a = list(map(int,input().split()))
plus, minus, multiply, divide = map(int, input().split())
data = []
for i in range(plus):
    data.append('+')
for i in range(minus):
    data.append('-')
for i in range(multiply):
    data.append('*')
for i in range(divide):
    data.append('/')
good = list(permutations(data,len(data)))
queue = deque(good)
max_value = -1e9
min_value = 1e9

while queue:
    init = a[0]
    yeonsanja = queue.popleft()
    o = 1
    for i in yeonsanja:
        if i =='+':
            init += a[o]
        elif i =='-':
            init -= a[o]
        elif i =='*':
            init *= a[o]
        else :
            init /= a[o]
            init = int(init)
        o += 1
    if init > max_value:
        max_value = init
    if init < min_value:
        min_value = init

print(max_value, min_value)
```

```python
n = int(input())
data = list(map(int,input().split()))
add, sub, mul, div = map(int, input().split())

min_value = 1e9
max_value = -1e9

def dfs(i, now):
    global min_value, max_value, add, sub, mul, div
    if i ==n:
        min_value = min(min_value, now)
        max_value = max(max_value, now)
    else:
        if add >0:
            add -= 1
            dfs(i + 1, now + data[i])
            add += 1
        if sub>0:
            sub -= 1
            dfs(i + 1, now - data[i])
            sub += 1
            
        if mul > 0:
            mul -= 1
            dfs(i+1, now *data[i])
            mul += 1
            
        if div >0:
            div -=1
            dfs(i+1, int(now/data[i]))
            div +=1
            
dfs(1,data[0])
print(max_value)
print(min_value)
        
```

```python
from itertools import combinations
from collections import deque
n = int(input())
board = []
teachers = []
spaces = []

for i in range(n):
    board.append(list(input().split()))
    for j in range(n):
        if board[i][j] == 'T':
            teachers.append((i,j))
        if board[i][j] =='X':
            spaces.append((i,j))

def watch(x,y,direction):
    if direction == 0:
        while y >= 0:
            if board[x][y] =='S':
                return True
            if board[x][y]=='O':
                return False
            y-=1

    if direction == 1:
        while y<n:
            if board[x][y] =='S':
                return True
            if board[x][y]=='O':
                return False
            y += 1

    if direction ==2:
        while x>=0:
            if board[x][y] =='S':
                return True
            if board[x][y]=='O':
                return False
            x-=1

    if direction ==3:
        while x<n:
            if board[x][y] =='S':
                return True
            if board[x][y]=='O':
                return False
            x+=1
            
    return False

def process():
    for x,y in teachers:
        for i in range(4):
            if watch(x,y,i):
                return True
    return False

find = False

for data in combinations(spaces,3):
    for x, y in data:
        board[x][y] ='O'
    if not process():
        find = True
        break
    for x, y in data:
        board[x][y] = 'X'


```

# 인구이동 문제(2022.07.20)

>**간단한 리뷰**

좋았던 점:

bfs에 대한 이해도는 상당히 높아졌다.

어려웠던 점:

각 국이 연결 되어있지 않다는 점을 있는 그대로 구현하려다보니 상당히 헷갈렸다. 이 부분은 타인의 코드를 참조했는데, 있는 그대로 모두 구현하는 것보다는 창의적인 아이디어가 필요하다는 것을 느낌






```python
from collections import deque

N, L, R = map(int, input().split())

data = []
for _ in range(N):
    data.append(list(map(int, input().split())))

dx = [-1, 0, 1, 0]
dy = [0, 1, 0, -1]

union = [ [-1]* N for _ in range(N)]

def search(x, y, index):
    dq = deque()
    united = [] 
    united.append((x, y))
    dq.append((x, y))
    summary = data[x][y]
    union[x][y] = index
    count = 1
    while dq:
        x, y = dq.popleft()
        for i in range(4):
            nx = x + dx[i]
            ny = y + dy[i]
            if 0 <= nx < N and 0 <= ny < N and union[nx][ny] == -1:
                if L <= abs(data[x][y] - data[nx][ny]) <= R:
                    united.append((nx, ny))
                    dq.append((nx, ny))
                    union[nx][ny] = index
                    summary += data[nx][ny]
                    count += 1

    for r, c in united:
        data[r][c] = summary // count

    return

answer = 0
# 이부분 헷갈림
while True:
    union = [[-1] * N for _ in range(N)]
    index = 0
    for i in range(N):
        for j in range(N):
            if union[i][j] == -1:
                search(i, j, index)
                index += 1

    if index == N * N:
        break
    answer += 1

print(answer)
```

# 블록 이동하기(2022.07.20)

>**간단한 리뷰**

좋았던 점:

어려운 문제에 접근했다는 점에서 칭찬한다.

어려웠던 점:

알고리즘은 외우는 것이 아닌 응용의 연속이라는 것을 뼈저리 느꼈다. 3시간동안 삽질을 하였고, 도저히 풀지 못해서 답지를 확인했는데 방향성은 얼추 맞았지만 신경쓰지 못한 부분들이 굉장히 많이 있다는 것을 확인 할 수 있었다. 다양한 생각을 통한 다방향의 접근성을 가진 빅데이터 분석가가 돼야겠다.

배운 점:

최소거리 및 산발적은 bfs 사용하는 것이 맞음


```python
from collections import deque

def get_next_pos(pos, board):
    next_pos = []
    pos = list(pos)
    pos1_x, pos1_y, pos2_x, pos2_y = pos[0][0], pos[0][1], pos[1][0], pos[1][1]
    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]
    for i in range(4):
        pos1_next_x, pos1_next_y, pos2_next_x, pos2_next_y = pos1_x + dx[i], pos1_y + dy[i], pos2_x + dx[i], pos2_y + dy[i]
        if board[pos1_next_x][pos1_next_y] == 0 and board[pos2_next_x][pos2_next_y] ==0:
            next_pos.append({(pos1_next_x, pos1_next_y), (pos2_next_x, pos2_next_y)})
    if pos1_x == pos2_x:
        for i in [-1, 1]:
            if board[pos1_x +i][pos1_y] ==0 and board[pos2_x +i][pos2_y] ==0:
                next_pos.append({(pos1_x, pos1_y),(pos1_x +i, pos1_y)})
                next_pos.append({(pos2_x, pos2_y),(pos2_x +i, pos2_y)})
    elif pos1_y == pos2_y:
        for i in [-1,1]:
            if board[pos1_x][pos1_y + i] ==0 and board[pos2_x][pos2_y +i] == 0:
                next_pos.append({(pos1_x, pos1_y),(pos1_x, pos1_y +i)})
                next_pos.append({(pos2_x, pos2_y),(pos2_x, pos2_y +i)})
    return next_pos

def solution(board):
    n = len(board)
    new_board = [[1] * (n+2) for _ in range(n+2)]
    for i in range(n):
        for j in range(n):
            new_board[i + 1][j + 1] = board[i][j]
    q = deque()
    visited = []
    pos = {(1,1), (1,2)}
    q.append((pos, 0 ))
    visited.append(pos)
    
    while q:
        pos, cost = q.popleft()
        if (n,n) in pos:
            return cost
        for next_pos in get_next_pos(pos, new_board):
            if next_pos not in visited:
                q.append((next_pos, cost + 1))
                visited.append(next_pos)
    return 0
```
