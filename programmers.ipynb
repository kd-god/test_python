```python
def solution(record):
    answer = []
    values = []
    keys = []
    for i in record:
        new_record = i.split()
        if len(new_record) == 3:
            values.append(new_record[2])
            keys.append(new_record[1])
    my_dict = dict(zip(keys,values))  
    
    for i in record:
        if i.split()[0] == 'Enter':
            answer.append(my_dict[i.split()[1]]+'님이 들어왔습니다.')
        elif i.split()[0] == 'Leave':
            answer.append(my_dict[i.split()[1]]+ '님이 나가셨습니다.')
        else:
            continue
    print(answer)
solution(record)
```

```python
def solution(w,h):
    answer = 1
    target = w * h
    array = [w, h]
    
    for i in range(min(array), 0, -1):
        if w % i == 0 and h % i ==0:
            answer = i
            break
            
    answer = target -(w+h-i) 
    return answer
```
