```python
def solution(record):
    answer = []
    values = []
    keys = []
    for i in record:
        new_record = i.split()
        if len(new_record) == 3:
            values.append(new_record[2])
            keys.append(new_record[1])
    my_dict = dict(zip(keys,values))  
    
    for i in record:
        if i.split()[0] == 'Enter':
            answer.append(my_dict[i.split()[1]]+'님이 들어왔습니다.')
        elif i.split()[0] == 'Leave':
            answer.append(my_dict[i.split()[1]]+ '님이 나가셨습니다.')
        else:
            continue
    print(answer)
solution(record)
```

```python
def solution(w,h):
    answer = 1
    target = w * h
    array = [w, h]
    
    for i in range(min(array), 0, -1):
        if w % i == 0 and h % i ==0:
            answer = i
            break
            
    answer = target -(w+h-i) 
    return answer
```

```python
from itertools import product

def solution(n):
    answer = []
    for i in range(1, n + 1):
        for j in product([1,2,4], repeat = i):
            if len(j) > i+1:
                continue
            else:
                a = ''
                for k,v in enumerate(j):
                    a += str(j[k])
                answer.append(a)
    return answer[n-1]
```

```python
def solution(n):
    answer_array = ['1', '2', '4']
    answer = ''
    while n:
        n-=1
        answer =answer_array[n%3] + answer 
        n//=3
    return answer
```
